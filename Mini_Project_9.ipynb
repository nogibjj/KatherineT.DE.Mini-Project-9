# Data Engineering Mini Project 9
## Import Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Read Dataset

```python
dataset = "https://raw.githubusercontent.com/nogibjj/KatherineT.DE.Mini-Project_9/main/cereal.csv"
cereal = pd.read_csv(dataset,sep=";")
# Glance data
cereal.head()
```

## Data Manipulation


```python
# summary statistics function
cereal.describe()
```

## Data Visualization


```python
plt.hist(cereal["calories"], bins=10)
plt.title("Cereak Calories Histogram")
plt.xlabel("Calories")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(cereal["sugars"],cereal['calories'],alpha=0.5)
plt.xlabel("Sugar")
plt.ylabel("Calories")
plt.title("Cereal Sugar vs. Calories")
plt.savefig("Cereal_Scatter.png")
plt.show()
```
